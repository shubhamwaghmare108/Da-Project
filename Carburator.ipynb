```python
import random
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta

def generate_random_contract_dates(
    start_year_min=2016,
    start_year_max=2025,
    duration_months_min=6,
    duration_months_max=36
):
    """
    Generates a random contract start date and calculates a subsequent end date.

    Args:
        start_year_min (int): The earliest year for the start date.
        start_year_max (int): The latest year for the start date.
        duration_months_min (int): The minimum contract duration in months.
        duration_months_max (int): The maximum contract duration in months.

    Returns:
        tuple: A tuple containing (start_date, end_date) as datetime objects.
    """
    # 1. Generate a random Start Date
    
    # Define the range for the start date generation (e.g., between 2020-01-01 and 2025-12-31)
    date_range_start = datetime(start_year_min, 1, 1)
    date_range_end = datetime(start_year_max, 12, 31)

    # Calculate the total number of days in the range
    time_difference = date_range_end - date_range_start
    days_in_range = time_difference.days

    # Pick a random number of days to offset the start date
    random_days = random.randrange(days_in_range)
    
    # Calculate the final Start Date
    start_date = date_range_start + timedelta(days=random_days)

    # 2. Generate a random Duration (in months)
    
    # Randomly select a contract duration between the min and max months
    contract_duration_months = random.randint(duration_months_min, duration_months_max)

    # 3. Calculate the End Date
    
    # Add the random duration to the start date using relativedelta for accurate month addition
    # Note: You must install python-dateutil for relativedelta: pip install python-dateutil
    end_date = start_date + relativedelta(months=contract_duration_months)

    return start_date, end_date

# --- Example Usage ---
# Generate the dates
contract_start, contract_end = generate_random_contract_dates()

# Print the results in a standard format
print(f"Contract Start Date: {contract_start.strftime('%Y-%m-%d')}")
print(f"Contract Duration: {((contract_end - contract_start).days // 30.4375)} months (approx)") # Display approx duration
print(f"Contract End Date: {contract_end.strftime('%Y-%m-%d')}")
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta

# --- 1. Redefine the function (for a complete, runnable block) ---
def generate_random_contract_dates(
    start_year_min=2015,
    start_year_max=2024,
    duration_months_min=13,
    duration_months_max=36
):
    date_range_start = datetime(start_year_min, 1, 1)
    date_range_end = datetime(start_year_max, 12, 31)
    time_difference = date_range_end - date_range_start
    days_in_range = time_difference.days
    random_days = random.randrange(days_in_range)
    start_date = date_range_start + timedelta(days=random_days)
    contract_duration_months = random.randint(duration_months_min, duration_months_max)
    end_date = start_date + relativedelta(months=contract_duration_months)
    # Return datetime objects for better DataFrame use (can convert later if needed)
    return start_date, end_date

# --- 2. Generate the List of Data ---
contract_list = [generate_random_contract_dates() for i in range(1990)]

# --- 3. Create the DataFrame ---
df_contracts = pd.DataFrame(
    contract_list, 
    columns=['ContractStartDate', 'ContractEndDate']
)

# --- 4. Verify Data Types (Optional but Recommended) ---
# Ensure the columns are in the datetime format for easy manipulation
df_contracts['ContractStartDate'] = pd.to_datetime(df_contracts['ContractStartDate'])
df_contracts['ContractEndDate'] = pd.to_datetime(df_contracts['ContractEndDate'])

print("DataFrame Head:")
print(df_contracts.head())
print("\nDataFrame Info:")
print(df_contracts.info())
```
